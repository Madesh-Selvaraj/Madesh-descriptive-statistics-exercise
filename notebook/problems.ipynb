# Descriptive statistics problems

### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
# TODO
import numpy as np 
import matplotlib.pyplot as plt

np.random.seed(42)

#Array following normal distribution 
normal_array = np.random.normal(loc=0, scale=1, size=100)
print(f"normal distribution array: {normal_array}")

#Array following Chi-square distribution 
chi_square_array = np.random.chisquare(df=3, size=100)
print(f"chi-square distribution: {chi_square_array}")
````

````python
# Measures of Central Tendency 
import numpy as np 
import statistics as stats
import matplotlib.pyplot as plt

# Normal Distribution 
print("Normal Distribution central tendency")
mean = stats.mean(normal_array)
print(f"mean: {mean:.4f}") # Mean 
median = stats.median(normal_array)
print(f"median: {median:.4f}") # Median 
mode = stats.mode(normal_array)
print(f"mode: {mode:.4f}") # Mode

plt.hist(normal_array, bins=10, alpha=0.8)

plt.axvline(mean, color='r', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(median, color='g', linestyle='dashed', linewidth=2, label='Median')
plt.axvline(mode, color='b', linestyle='dashed', linewidth=2, label='Mode')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Mean, Median, and Mode')
plt.legend()

plt.show() 

# Chi-square Distribution 
print("Chi-square Distribution central tendency")
mean = stats.mean(chi_square_array)
print(f"mean: {mean:.4f}") # Mean 
median = stats.median(chi_square_array)
print(f"median: {median:.4f}") # Median 
mode = stats.mode(chi_square_array)
print(f"mode: {mode:.4f}") # Mode

plt.hist(chi_square_array, bins=10, alpha=0.8)

plt.axvline(mean, color='r', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(median, color='g', linestyle='dashed', linewidth=2, label='Median')
plt.axvline(mode, color='b', linestyle='dashed', linewidth=2, label='Mode')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Mean, Median, and Mode')
plt.legend()

plt.show() 
````

````python
# Measures of Dispersion 
import numpy as np
import matplotlib.pyplot as plt

# Normal Distribution 
print("Normal Distribution Dispersion")
normal_range = max(normal_array) - min(normal_array)
print(f"range: {normal_range:.4f}")                 # Range of Normal Distribution

normal_mean = stats.mean(normal_array)   
print(f"mean: {normal_mean:.4f}") # Mean                            # Mean of Normal Distribution 
normal_variance = stats.variance(normal_array)                      # Variance of Normal Distribution                             
normal_std_dev = stats.stdev(normal_array)                          # Standard Deviation of Normal Distribution
print(f"variance: {normal_variance:.4f} and std dev: {normal_std_dev:.4f}")

fig, ax = plt.subplots()

ax.hist(normal_array, bins=20, density=True, alpha=0.6, color='blue')

ax.axvline(normal_mean, color='red', linestyle='dashed', linewidth=2, label='Mean')
ax.axvline(normal_mean + normal_std_dev, color='green', linestyle='dashed', linewidth=2, label='Std Dev')
ax.axvline(normal_mean - normal_std_dev, color='green', linestyle='dashed', linewidth=2)
ax.axvline(min(normal_array), color='black', linestyle='dashed', linewidth=2, label='Range')
ax.axvline(max(normal_array), color='black', linestyle='dashed', linewidth=2)


ax.set_title('Normal Data Dispersion')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.legend()


plt.show()


# Chi-square Distribution 
print("Chi-square Distribution Dispersion")
chi_square_range = max(chi_square_array) - min(chi_square_array)
print(f"range: {chi_square_range:.4f}")         # Range of Chi-square Distribution

chi_square_mean = stats.mean(chi_square_array)   
print(f"mean: {normal_mean:.4f}") # Mean                            # Mean of Chi-square Distribution 
chi_square_variance = stats.variance(chi_square_array)              # Variance of Chi-square Distribution
chi_square_std_dev = stats.stdev(chi_square_array)                  # Standard Deviation of Chi-square Distribution
print(f"variance: {chi_square_variance:.4f} and std dev: {chi_square_std_dev:.4f}")

fig, ax = plt.subplots()

ax.hist(chi_square_array, bins=20, density=True, alpha=0.6, color='blue')

ax.axvline(chi_square_mean, color='red', linestyle='dashed', linewidth=2, label='Mean')
ax.axvline(chi_square_mean + chi_square_std_dev, color='green', linestyle='dashed', linewidth=2, label='Std Dev')
ax.axvline(chi_square_mean - chi_square_std_dev, color='green', linestyle='dashed', linewidth=2)
ax.axvline(min(chi_square_array), color='black', linestyle='dashed', linewidth=2, label='Range')
ax.axvline(max(chi_square_array), color='black', linestyle='dashed', linewidth=2)

ax.set_title('Chi-square Data Dispersion')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.legend()

plt.show()
````

````python
# Shape Measures 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import skew
from scipy.stats import kurtosis

#Skewness 
skew_normal = skew(normal_array)
print(f"Normal skewness: {skew_normal:.4f}")
skew_chi = skew(chi_square_array)
print(f"Chi skewness: {skew_chi:.4f}")

# Kurtosis 
kurt_normal = kurtosis(normal_array)
print(f"Normal kurtosis: {kurt_normal:.4f}")
kurt_chi = kurtosis(chi_square_array)
print(f"Chi kurtosis: {kurt_chi:.4f}")
````

Normal distribution has a skewness close to 0 with mean/median/mode are close to 0. Kurtosis is also close to 0(mesokurtic). So the vector follows a normal distribution with high accuracy! 

Chi-square distribution has a skewness close to 1, having positive skew (large amount of data on left). Kurtosis > 3 (leptokurtic). These matches exactly with a chi-squared distribution with mean larger than mode/median.

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
# TODO
import math 
# Mean Calculation 
def mean_function(number_list): 
  return sum(number_list) / len(number_list)
#Std Dev calculation 
def std_function(number_list): 
  mean = mean_function(number_list)
  variance = sum((x - mean) ** 2 for x in number_list) / len(number_list)
  return variance ** 0.5 
#Output
number_list = [4, 2, 5, 8 , 6]
print(f"Sample list is: {number_list}")
print(f"Mean of the sample list is {mean_function(number_list)}")
print(f"Standard deviation of the smaple list is : {std_function(number_list)}")
````
